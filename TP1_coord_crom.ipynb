```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv.imread('TP1/CoordCrom_1.png')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        pixel = img[i, j].copy()
        img[i, j, 0]=(pixel[0]/pixel.sum())*255
        img[i, j, 1]=(pixel[1]/pixel.sum())*255
        img[i, j, 2]=(pixel[2]/pixel.sum())*255
plt.imshow(img)
plt.show()
```

```python
img = cv.imread('TP1/WP_B.png')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

R=np.amax(img[:,:,0])
G=np.amax(img[:,:,1])
B=np.amax(img[:,:,2])
print(R,G,B)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        pixel = img[i, j].copy()
        img[i, j, 0]=(255/R)*pixel[0]
        img[i, j, 1]=(255/G)*pixel[1]
        img[i, j, 2]=(255/B)*pixel[2]

plt.imshow(img)
plt.show()
```
